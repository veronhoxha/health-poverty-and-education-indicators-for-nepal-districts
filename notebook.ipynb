### Imports

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

### Basic Exploration

```python
gdf = gpd.read_file("data/nepal.shp")
```

```python
print(gdf.info())
```

```python
gdf.plot()
```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))

gdf.plot(ax=ax, column="povindex", cmap="Reds", legend=True, 
        legend_kwds={"label": "Poverty Index", "orientation": "horizontal"})
ax.set_title("Poverty Index in Nepal")
ax.set_axis_off()

plt.show()
```
